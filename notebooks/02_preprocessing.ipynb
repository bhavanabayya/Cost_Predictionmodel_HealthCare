```python
import pandas as pd
import os
```

```python
# loading the dataset

# Construct path
data_path = os.path.join("..", "data", "insurance.csv")
df = pd.read_csv(data_path)

df.head()
```

```python
# Lets do some one hot encoding of sex, region and smoker columns
df_encoded = pd.get_dummies(df, columns=['sex', 'region', 'smoker'], drop_first=True)
df_encoded = df_encoded.astype(int)

df_encoded.head()
```

```python
# Save the cleaned dataset
df_encoded.to_csv("../data/insurance_cleaned.csv", index=False)
```
